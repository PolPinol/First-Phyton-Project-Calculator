```python
#Calculadora

from tkinter import *
from PIL import ImageTk, Image
from decimal import Decimal

raiz=Tk()
raiz.title("Ventana")
raiz.resizable(True,True)
raiz.config(bg="black")

miFrame=Frame()
miFrame.pack()
miFrame.config(bg="gray")
miFrame.config(width="800", height="800")


memoria=StringVar()

operacio=StringVar()
operacio.set("0")

numero=StringVar()
numero.set("0")

lectura=StringVar()
lectura.set("0")

def posarNumero(n):
    if operacio.get() == "=":
        numero.set("0")
        operacio.set("0")
    if "." in numero.get() and n == ".":
        pass
    elif n == ".":
        numero.set(numero.get()+n)
    elif 0 == float(n) and numero.get()=="0":
        pass
    elif numero.get()=="0":
        numero.set(n)
    else:
        numero.set(numero.get()+n)

def borrarC():
    memoria.set("0")
    numero.set("0")
    lectura.set("0")
    operacio.set("0")
        
def borrarCE():
    numero.set("0")

def dividir2():
    a=float(numero.get())
    if a==0:
        pass
    else:
        if round(a)==a:
            b=format(a, '.0f')
            lectura.set("1/("+b+")=")
        else:
            lectura.set("1/("+str(a)+")=")
        if round(1/a)==1/a:
            c=format(1/a, '.0f') 
            numero.set(c)
            memoria.set(c)
        else:
            numero.set(1/a)
            memoria.set(1/a)

def elevar2():
    a=float(numero.get())
    if round(a)==a:
        b=format(a, '.0f')
        lectura.set("("+b+")^2=")
    else:
        lectura.set("1/("+str(a)+")=")
    if round(a*a)==a*a:
        c=format(a*a, '.0f') 
        numero.set(c)
        memoria.set(c)
    else:
        numero.set(a*a)
        memoria.set(a*a)
        
    
def operar(op):
    if operacio.get() == "0":
        pass
    elif operacio.get() == "=":
        pass
    else:
        equal("no")
    memoria.set(numero.get())
    lectura.set(memoria.get()+op)
    numero.set("0")
    operacio.set(op)
    
def equal(end):
    if operacio.get() == "+" or operacio.get() == "-" or operacio.get() == "*" or operacio.get() == "/" :
        lectura.set(memoria.get()+operacio.get()+numero.get()+"=")
        a=float(memoria.get())
        b=float(numero.get())
        
        if operacio.get()=="+":
            c=a+b
        elif operacio.get()=="-":
            c=a-b
        elif operacio.get()=="*":
            c=a*b
        elif operacio.get()=="/":
            c=a/b
            
        if round(c)==c:
            c=format(c, '.0f')
        numero.set(c)
        if end=="yes":
            operacio.set("=")
        
        
        
        
### Fila 0

cuadroAns=Entry(miFrame, textvariable=lectura, justify=RIGHT)
cuadroAns.grid(row=0, column=1, columnspan=4, pady="3",padx="3")

cuadro=Entry(miFrame, textvariable=numero, justify=RIGHT)
cuadro.grid(row=1, column=1, columnspan=4, pady="3",padx="3")


#### Fila 1

boton1=Button(miFrame, text="%", width="5", height="2")
boton1.grid(row=2, column=1, pady="3",padx="3")

boton2=Button(miFrame, text="CE", width="5", height="2", command=lambda:borrarCE())
boton2.grid(row=2, column=2, pady="3",padx="3")

boton3=Button(miFrame, text="C", width="5", height="2", command=lambda:borrarC())
boton3.grid(row=2, column=3, pady="3",padx="3")

boton4=Button(miFrame, text="<x", width="5", height="2")
boton4.grid(row=2, column=4, pady="3",padx="3")


#### Fila 2

boton5=Button(miFrame, text="1/x", width="5", height="2", command=lambda:dividir2())
boton5.grid(row=3, column=1, pady="3",padx="3")

boton6=Button(miFrame, text="x^2", width="5", height="2", command=lambda:elevar2())
boton6.grid(row=3, column=2, pady="3",padx="3")

boton7=Button(miFrame, text="sqrt(x)", width="5", height="2")
boton7.grid(row=3, column=3, pady="3",padx="3")

boton8=Button(miFrame, text="/", width="5", height="2", command=lambda:operar("/"))
boton8.grid(row=3, column=4, pady="3",padx="3")


#### Fila 3

boton9=Button(miFrame, text="7", width="5", height="2", command=lambda:posarNumero("7"))
boton9.grid(row=4, column=1, pady="3",padx="3")

boton10=Button(miFrame, text="8", width="5", height="2", command=lambda:posarNumero("8"))
boton10.grid(row=4, column=2, pady="3",padx="3")

boton11=Button(miFrame, text="9", width="5", height="2", command=lambda:posarNumero("9"))
boton11.grid(row=4, column=3, pady="3",padx="3")

boton12=Button(miFrame, text="*", width="5", height="2", command=lambda:operar("*"))
boton12.grid(row=4, column=4, pady="3",padx="3")



#### Fila 4

boton13=Button(miFrame, text="4", width="5", height="2", command=lambda:posarNumero("4"))
boton13.grid(row=5, column=1, pady="3",padx="3")

boton14=Button(miFrame, text="5", width="5", height="2", command=lambda:posarNumero("5"))
boton14.grid(row=5, column=2, pady="3",padx="3")

boton15=Button(miFrame, text="6", width="5", height="2", command=lambda:posarNumero("6"))
boton15.grid(row=5, column=3, pady="3",padx="3")

boton16=Button(miFrame, text="-", width="5", height="2", command=lambda:operar("-"))
boton16.grid(row=5, column=4, pady="3",padx="3")



#### Fila 5

boton17=Button(miFrame, text="1", width="5", height="2", command=lambda:posarNumero("1"))
boton17.grid(row=6, column=1, pady="3",padx="3")

boton18=Button(miFrame, text="2", width="5", height="2", command=lambda:posarNumero("2"))
boton18.grid(row=6, column=2, pady="3",padx="3")

boton19=Button(miFrame, text="3", width="5", height="2", command=lambda:posarNumero("3"))
boton19.grid(row=6, column=3, pady="3",padx="3")

boton20=Button(miFrame, text="+", width="5", height="2", command=lambda:operar("+"))
boton20.grid(row=6, column=4, pady="3",padx="3")


#### Fila 6

boton21=Button(miFrame, text="+/-", width="5", height="2")
boton21.grid(row=7, column=1, pady="3",padx="3")

boton22=Button(miFrame, text="0", width="5", height="2", command=lambda:posarNumero("0"))
boton22.grid(row=7, column=2, pady="3",padx="3")

boton23=Button(miFrame, text=",", width="5", height="2", command=lambda:posarNumero("."))
boton23.grid(row=7, column=3, pady="3",padx="3")

boton24=Button(miFrame, text="=", width="5", height="2", command=lambda:equal("yes"))
boton24.grid(row=7, column=4, pady="3",padx="3")




raiz.mainloop()
```
